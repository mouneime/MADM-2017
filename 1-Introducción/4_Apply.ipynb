### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def mayoria_edad(x):
    if x > 17:
        return True
    else:
        return False
```

```python
estudianes_alcohol['legal_drinker'] = estudianes_alcohol['age'].apply(mayoria_edad)
estudianes_alcohol.head()
```
